## Notebook Workflows

This notebook will serve as the starting point for the others. That is, it will make the calls to the other notebooks.

```
dbutils.notebook.help("run")
```

Run notebook with just passing notebook name and timeout seconds.

```
dbutils.notebook.run("Notebook Workflows - Print Output", timeout_seconds=5)
```

Run notebook to run excuting notebook that has a widget.

```
dbutils.notebook.run("Notebook Workflows - Pass Arguments", timeout_seconds=5, arguments={"timeout_seconds": "10"})
```

If no arguments are passed it can run without passing arguments and let the notebook use the default value.

```
dbutils.notebook.run("Notebook Workflows - Pass Arguments", timeout_seconds=5)
```

Return results of a notebook.

```
result = dbutils.notebook.run("Notebook Workflows - Return Single Result", timeout_seconds=5)
result
```

Return temp table name

```
data = dbutils.notebook.run("Notebook Workflow - Return Temp Table", timeout_seconds=5)
```

```
global_db = spark.conf.get("spark.sql.globalTempDatabase")
```

```
data_table = "{0}.{1}".format(global_db, data)
```

```
table_display = table(data_table)
```

```
display(table_display)
```

Return JSON data

```
dbutils.notebook.run("Notebook Workflow - Return JSON", timeout_seconds=5)
```

Use DBFS to pass data between notebooks

```
data_path = dbutils.notebook.run("Notebook Workflows - DBFS Data", timeout_seconds=5)
```

```
df = spark.read.csv(data_path, header=True)
```

```
display(df)
```
