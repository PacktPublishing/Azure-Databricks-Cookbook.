```
storage_account_name = "<Azure Blob Storage Account Name>"
storage_account_key = "<Azure Blob Storage Account Key>"
container = "<Container Name>"

spark.conf.set("fs.azure.account.key.{0}.blob.core.windows.net".format(storage_account_name), storage_account_key)
```

```
df = spark.read \
  .option("header", "true") \
  .option("inferSchema", "true") \
  .option("delimiter", ";") \
  .csv("wasbs://{0}@{1}.blob.core.windows.net/bank.csv".format(container, storage_account_name))
```

```
display(df.agg({"balance": "avg"}))
```

```
display(df.groupBy().avg("balance"))
```
