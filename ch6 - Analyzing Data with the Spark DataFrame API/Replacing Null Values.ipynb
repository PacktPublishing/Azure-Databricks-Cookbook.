```
df = spark.read \
  .option("header", "true") \
  .option("inferSchema", "true") \
  .option("delimiter", ",") \
  .csv("/databricks-datasets/Rdatasets/data-001/csv/datasets/airquality.csv")
```

```
display(df)
```

```
df = df.select(["Ozone"])
display(df)
```

```
from pyspark.sql.functions import when

df = df.withColumn("Ozone", when(df["Ozone"] == 'NA', None).otherwise(df["Ozone"]))
```

```
display(df)
```

```
df.count() - df.na.drop().count()
```

```
ozone_mean = df.agg({"Ozone": "mean"}).collect()[0][0]

display(df.na.fill({"Ozone": int(ozone_mean)}))
```
