```
from graphframes import GraphFrame
```

```
verticies = spark.createDataFrame([(1, "chair", 50.00), (2, "table", 99.00), (3, "pen", 5.00), (4, "server", 999.00), (5, "desk", 70.00)], ["id", "product", "price"])
```

```
edges = spark.createDataFrame([(1, 2, "home"), (1, 3, "home"), (2, 4, "office"), (3, 5, "office")], ["src", "dst", "relationship"])
```

```
graph = GraphFrame(verticies, edges)
```

```
display(graph.vertices)
```

```
display(graph.edges)
```
