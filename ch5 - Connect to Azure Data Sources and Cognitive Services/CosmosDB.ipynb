```
df = spark.createDataFrame([(22, "Jim")], ["age", "name"])
```

```
df.write \
 .format("com.microsoft.azure.cosmosdb.spark") \
 .option("Endpoint", "<CosmosDB Endpoint URL>") \
 .option("Masterkey", "<CosmosDB Master key>") \
 .option("Database", "<CosmosDB Database>") \
 .option("Collection", "<CosmosDB collection>") \
 .option("Upsert", "true") \
 .mode("append") \
 .save()
```
