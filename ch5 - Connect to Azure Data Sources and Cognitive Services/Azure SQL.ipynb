The below data frame assumes there's a table created with the following schema:
 - Age - integer
 - FirstName - string

```
data = spark.createDataFrame([(22, "John")], ["Age", "FirstName"])
```

```
data.show()
```

```
data.write.jdbc(connectionString, "<TableName>", mode='append')
```

### Separate way to read data from a JDBC connection string.

```
new_df = spark.read.format("jdbc") \
  .option("url", connectionString) \
  .option("dbtable", "<TableName>") \
  .load()
```
