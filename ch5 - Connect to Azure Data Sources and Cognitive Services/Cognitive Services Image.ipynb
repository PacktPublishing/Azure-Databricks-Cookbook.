```
vision_key = "<Vision Key>"
vision_endpoint = "<Vision Endpoint URL>"
```

```
from azure.cognitiveservices.vision.computervision import ComputerVisionClient
from azure.cognitiveservices.vision.computervision.models import VisualFeatureTypes
from msrest.authentication import CognitiveServicesCredentials
```

```
credentials = CognitiveServicesCredentials(vision_key)
client = ComputerVisionClient(vision_endpoint, credentials)
```

```
image_url = "https://upload.wikimedia.org/wikipedia/commons/7/7d/Bigben.jpg"

image_analysis = client.analyze_image(image_url, visual_features=[VisualFeatureTypes.tags, VisualFeatureTypes.categories, VisualFeatureTypes.objects, VisualFeatureTypes.adult])

for tag in image_analysis.tags:
  print(tag)
```

```
for category in image_analysis.categories:
  print(category)
```

```
for obj in image_analysis.objects:
  print(obj.object_property, obj.confidence)
```

```
print("Is adult content?", image_analysis.adult.is_adult_content)
print("Is racy content?", image_analysis.adult.is_racy_content)
print("Adult score", image_analysis.adult.adult_score)
print("Racy score", image_analysis.adult.racy_score)
```

```
description = client.describe_image(image_url)

for caption in description.captions:
  print(caption.text, caption.confidence)
```
