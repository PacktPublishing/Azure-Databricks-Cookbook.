```
redis_host = "<Redis Server URL>"
redis_port = 6380
redis_pw = "<Redis Password>"
```

```
import redis
cache = redis.StrictRedis(host=redis_host, port=redis_port, password=redis_pw, ssl=True)
```

```
cache.set("language", "Python")
```

```
cache.get("language")
```

```
for key in cache.scan_iter():
  print(key, cache.get(key))
```

```
for key in cache.scan_iter(match="l*"):
  print(key, cache.get(key))
```

```
item = {"key1": "value1", "key2": "value2"}

cache.hmset("hmKey", item)
```

```
cache.hgetall("hmKey")
```
