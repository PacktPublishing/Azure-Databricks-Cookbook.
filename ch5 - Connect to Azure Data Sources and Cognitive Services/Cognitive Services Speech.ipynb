```
dbutils.library.installPyPI("azure-cognitiveservices-speech")
```

```
import azure.cognitiveservices.speech as speech
```

```
speech_key = "<Speech Key>"
speech_region = "<Speech Region>"
```

```
audio_file = "/dbfs/mnt/audio/cookbook.wav"
```

```
speech_config = speech.SpeechConfig(subscription=speech_key, region=speech_region)
audio_config = speech.AudioConfig(filename=audio_file)
```

```
speech_client = speech.SpeechRecognizer(speech_config, audio_config=audio_config)
```

```
result = speech_client.recognize_once()
```
