```
connectionString = "<Event hub connection string>"

ehConf = {}
ehConf['eventhubs.connectionString'] = connectionString
```

```
df = spark \
  .readStream \
  .format("eventhubs") \
  .options(**ehConf) \
  .load()
```

```
df.writeStream \
  .format("delta") \
  .outputMode("append") \
  .option("checkpointLocation", "/delta/events/_checkpoints/event-hub") \
  .start("/delta/event-hub")
```
